```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
data = boston.data
data.shape
```

```python
def normalization(nor_data):
    for i in range(nor_data.shape[1]):
        minx = np.min(nor_data[:,i])
        maxx = np.max(nor_data[:,i])
        nor_data[:,i] = (nor_data[:,i]-minx)/(maxx-minx)
    return nor_data

def standardScaler(sds_data):
    for i in range(sds_data.shape[1]):
        mean = np.mean(sds_data[:,i])
        sd = np.std(sds_data[:,i])
        sds_data[:, i] = (sds_data[:, i]-mean)/sd
    return sds_data

data_nor = normalization(data)
# data_sds = standardScaler(data)
```

```python
y = boston.target
y = np.expand_dims(y, axis=1)
y.shape

# X = np.expand_dims(data_nor[:,0], axis=1)
X = data_nor[:]
```

```python
X.shape
```

# Initialize the value of bias and weight to zero

```python
def calculate_loss(y,yhat, t):
    return np.sum((y-yhat)**2)/t
```

```python
# Initialize bias
bias = 0

# Initialize weights
weights = np.zeros(X.shape[1])

# Initialize learning rate
L = 0.001

# Initialize number of epoch 
epochs = 2000 

# Calculate the total number of records
n = float(len(X))
```

```python
loss = []

# Performing Gradient Descent 
for i in range(epochs): 
    Y_pred = weights*X + bias  # The current predicted value of Y
    loss.append(calculate_loss(y, Y_pred, n))
    D_m = (-2/n) * sum(X * (y - Y_pred))  # Derivative wrt m
    D_c = (-2/n) * sum(y - Y_pred)  # Derivative wrt c
    weights = weights - L * D_m  # Update m
    bias = bias - L * D_c  # Update c
    
print (weights, bias)
```
